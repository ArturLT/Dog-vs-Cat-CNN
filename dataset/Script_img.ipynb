```python
import os
import shutil
from sklearn.model_selection import train_test_split

original_dataset_dir = 'D:/Workspace/jupyter/dataset/PetImages'  # caminho original
base_dir = 'D:/Workspace/jupyter/dataset'  # diretório base para onde os dados serão copiados

# Criar diretórios para treino, validação e teste
for split in ['train', 'val', 'test']:
    for category in ['cats', 'dogs']:
        dir_path = os.path.join(base_dir, split, category)
        os.makedirs(dir_path, exist_ok=True)
        print(f"Diretório criado: {dir_path}")

# Função para dividir e copiar os arquivos
def split_data(src_dir, category):
    images = [f for f in os.listdir(src_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    # Verificar se encontrou as imagens
    if len(images) == 0:
        print(f"[{category}] Nenhuma imagem encontrada no diretório {src_dir}")
        return

    print(f"[{category}] Total encontrado: {len(images)} imagens")
    
    # Divisão das imagens: 80% treino, 10% validação, 10% teste
    train, temp = train_test_split(images, test_size=0.2, random_state=42)
    val, test = train_test_split(temp, test_size=0.5, random_state=42)

    # Verificar o número de imagens após divisão
    print(f"[{category}] Distribuição de imagens:")
    print(f"  - Treinamento: {len(train)}")
    print(f"  - Validação: {len(val)}")
    print(f"  - Teste: {len(test)}")

    # Copiar para os diretórios apropriados
    for split_name, split_data in zip(['train', 'val', 'test'], [train, val, test]):
        for fname in split_data:
            src = os.path.join(src_dir, fname)
            dst = os.path.join(base_dir, split_name, category, fname)
            try:
                shutil.copyfile(src, dst)
                print(f"Imagem {fname} copiada para {split_name}/{category}")
            except Exception as e:
                print(f"Erro ao copiar {fname}: {e}")

# Copiar as imagens de gatos e cachorros
split_data(os.path.join(original_dataset_dir, 'Cat'), 'cats')
split_data(os.path.join(original_dataset_dir, 'Dog'), 'dogs')
```
