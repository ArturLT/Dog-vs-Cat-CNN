```python
import os, shutil
from sklearn.model_selection import train_test_split

# Caminho atualizado conforme sua organiza√ß√£o atual
original_dataset_dir = 'D:/Workspace/jupyter/dataset/PetImages'
base_dir = 'D:/Workspace/jupyter/dataset'  # onde a estrutura organizada vai ficar

# Criar diret√≥rios de sa√≠da (train, val, test)
for split in ['train', 'val', 'test']:
    for category in ['cats', 'dogs']:
        os.makedirs(os.path.join(base_dir, split, category), exist_ok=True)

# Fun√ß√£o para dividir e copiar os arquivos
def split_data(src_dir, category):
    images = [f for f in os.listdir(src_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    train, temp = train_test_split(images, test_size=0.2, random_state=42)
    val, test = train_test_split(temp, test_size=0.5, random_state=42)

    for split_name, split_data_list in zip(['train', 'val', 'test'], [train, val, test]):
        for fname in split_data_list:
            src = os.path.join(src_dir, fname)
            dst = os.path.join(base_dir, split_name, category, fname)
            try:
                shutil.copyfile(src, dst)
            except:
                pass  # ignora imagens corrompidas ou com erro

# Executar para as duas classes
split_data(os.path.join(original_dataset_dir, 'cats'), 'cats')
split_data(os.path.join(original_dataset_dir, 'dogs'), 'dogs')
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

original_dataset_dir = 'D:/Workspace/jupyter/dataset/PetImages'  # caminho original
base_dir = 'D:/Workspace/jupyter/dataset'  # diret√≥rio base para onde os dados ser√£o copiados

# Criar diret√≥rios para treino, valida√ß√£o e teste
for split in ['train', 'val', 'test']:
    for category in ['cats', 'dogs']:
        dir_path = os.path.join(base_dir, split, category)
        os.makedirs(dir_path, exist_ok=True)
        print(f"Diret√≥rio criado: {dir_path}")

# Fun√ß√£o para dividir e copiar os arquivos
def split_data(src_dir, category):
    images = [f for f in os.listdir(src_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    # Verificar se encontrou as imagens
    if len(images) == 0:
        print(f"[{category}] Nenhuma imagem encontrada no diret√≥rio {src_dir}")
        return

    print(f"[{category}] Total encontrado: {len(images)} imagens")
    
    # Divis√£o das imagens: 80% treino, 10% valida√ß√£o, 10% teste
    train, temp = train_test_split(images, test_size=0.2, random_state=42)
    val, test = train_test_split(temp, test_size=0.5, random_state=42)

    # Verificar o n√∫mero de imagens ap√≥s divis√£o
    print(f"[{category}] Distribui√ß√£o de imagens:")
    print(f"  - Treinamento: {len(train)}")
    print(f"  - Valida√ß√£o: {len(val)}")
    print(f"  - Teste: {len(test)}")

    # Copiar para os diret√≥rios apropriados
    for split_name, split_data in zip(['train', 'val', 'test'], [train, val, test]):
        for fname in split_data:
            src = os.path.join(src_dir, fname)
            dst = os.path.join(base_dir, split_name, category, fname)
            try:
                shutil.copyfile(src, dst)
                print(f"Imagem {fname} copiada para {split_name}/{category}")
            except Exception as e:
                print(f"Erro ao copiar {fname}: {e}")

# Copiar as imagens de gatos e cachorros
split_data(os.path.join(original_dataset_dir, 'Cat'), 'cats')
split_data(os.path.join(original_dataset_dir, 'Dog'), 'dogs')
```
